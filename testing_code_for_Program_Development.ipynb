```
import shelve
main = shelve.open('main.db')
db_list = list(main['db_list'].values())
  
i = 1
for dbname in db_list:
  print(1, '-', dbname)
  i += 1

db_load = input("Please choose a database to load. Enter its name. ")

while db_load not in db_list:
  print("Database not in the list. Try again")
  db_load = input("Please choose a database to load. Enter its name. ")
  
db = shelve.open(db_load+'.db')

return(db)


```
