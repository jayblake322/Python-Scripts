```
#@title Print Database List
def print_db_list():
  
  import shelve
  main = shelve.open('main.db')
  db_list = list(main['db_list'].values())
  
  i = 1
  for dbname in db_list:
    print(1, '-', dbname)
    i += 1
 
  db_load = input("Please choose a database to load. Enter its name. ")
  
  while db_load not in db_list:
    print("Database not in the list. Try again")
    db_load = input("Please choose a database to load. Enter its name. ")

  return(db_load)

```

```
#@title Teach Database


database = input("What database would you like to teach? ")

import shelve

numbers = shelve.open(database+'.db')

taught_set = input("Give the information you are about to teach a Reference: ")
lesson_material = input("OK. Please teach by entering the word and meaning separated by a '-': ")
  
dash_count = lesson_material.count('-') # check there is only one split
if dash_count == 1:
  term, meaning = lesson_material.split("-") # check there are two terms
  term = term.strip()
  meaning = meaning.strip()
else:
  print("Make sure you put in only one term and meaning with only one dash!")
  
#  add to the dictionary
material_dictionary = {}
material_dictionary[term] = meaning
  
# add finished dictionary to the database
numbers[taught_set] = material_dictionary
  
numbers.close()
```
