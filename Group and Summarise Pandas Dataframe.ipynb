```python
# Load an excel File
import pandas as pd

df = pd.read_excel('R:/Projects/Market Analysis and Database Records/Access Sales Data Databse Project/SC Sales Data/SC_All_Sales_Data.xlsx', sheet_name = 'Buildings')

df.head()
```

```python
selected_suburb = input("Please enter the suburb you would like some information on: ").title()
```

```python
suburb_index = df['Suburb'] == selected_suburb


suburb_df = df[suburb_index]
```

```python
suburb_df.sort_values(by=['First_Sale_Year'])

suburb_summary = suburb_df['First_Sale_Year'].value_counts()

print(suburb_summary.sort_index())
```
